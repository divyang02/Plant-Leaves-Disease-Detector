# model with all classes

## getting data

```python

batch_size = 64
num_classes = 38


name_to_labels = np.load('name_to_labels_all_classes.npy').item()
label_map_dict = {}
image_data_list = []
labels = []
```

```python

path = os.getcwd()
data_path = path + '/impTestcases'

for filename in os.listdir(data_path) :
    text_filename = filename.strip()
    if text_filename == '._.DS_Store' or text_filename == '.DS_Store' :
        continue
    for imagename in os.listdir(data_path + '/' + filename) :
        if imagename.startswith('.') :
            continue
        img_path = data_path + '/' + filename + '/' + imagename
        image_data_list.append(img_path)
        labels.append(name_to_labels[text_filename])
        
    label_map_dict[name_to_labels[text_filename]] = text_filename


print(len(image_data_list))
print(len(labels))
```

```python


import random

perm = list(range(len(image_data_list)))
random.shuffle(perm)
image_data_list = [image_data_list[index] for index in perm]
labels = [labels[index] for index in perm]
```

```python
test_data_list = []

for img_path in image_data_list :
    #print(img_path)
    img = image.load_img(img_path, target_size = (224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    test_data_list.append(x)

test_data = np.array(test_data_list)
test_data = np.rollaxis(test_data, 1, 0)
test_data = test_data[0]
print(test_data.shape)
```

```python
print(labels)
```

## loading model and predicting

```python
from keras.models import load_model

all_class_model = load_model('my_model.h5')
```

```python


#Showing predictions

predictions = all_class_model.predict(test_data,64)

count = 0
acc_avg = 0
for list_pred in predictions :
    if np.argmax(list_pred) == labels[count] :
        acc_avg += 1
    print('predicted = {} and label = {}'.format(np.argmax(list_pred), labels[count]))
    count += 1
print('accuracy = {}'.format((acc_avg/count)*100))

```

```python
print('accuracy is = {}%'.format((acc_avg/count)*100))
```

# Model with 2 classes

## getting data

```python
batch_size = 64
num_classes = 2

label_map_dict = {0 : 'healthy', 1 : 'diseased'}
image_data_list = []
labels = []
```

```python
#getting the data

path = os.getcwd()
data_path = path + '/impTestcases'

for filename in os.listdir(data_path) :
    text_filename = filename.strip()
    if text_filename == '.DS_Store' or text_filename == '._.DS_Store' :
        continue
    words = text_filename.split('___')
    for imagename in os.listdir(data_path + '/' + filename) :
        if imagename.startswith('.') :
            continue
        img_path = data_path + '/' + filename + '/' + imagename
        image_data_list.append(img_path)
        if words[len(words)-1] == 'healthy' :
            labels.append(0)
        else :
            labels.append(1)
            
    
print(len(image_data_list))
print(len(labels))

```

```python


import random

perm = list(range(len(image_data_list)))
random.shuffle(perm)
image_data_list = [image_data_list[index] for index in perm]
labels = [labels[index] for index in perm]
```

```python
test_data_list = []

for img_path in image_data_list :
    #print(img_path)
    img = image.load_img(img_path, target_size = (224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    test_data_list.append(x)

test_data = np.array(test_data_list)
test_data = np.rollaxis(test_data, 1, 0)
test_data = test_data[0]
print(test_data.shape)
```

```python
print(labels)
```

## loading model and predicting

```python
from keras.models import load_model

two_class_model = load_model('my_model_2class.h5')
```

```python
#Showing predictions

predictions = two_class_model.predict(test_data,64)

count = 0
acc_avg = 0
for list_pred in predictions :
    if np.argmax(list_pred) == labels[count] :
        acc_avg += 1
    print('predicted = {} and label = {}'.format(np.argmax(list_pred), labels[count]))
    count += 1
print('accuracy = {}'.format((acc_avg/count)*100))
```

```python
print('accuracy is = {}%'.format((acc_avg/count)*100))
```
